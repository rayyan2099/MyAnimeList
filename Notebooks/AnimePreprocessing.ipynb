```python
bad_rows = anime_df[(anime_df['status'] == '404') | (anime_df['status'] == '2195.0')].index
anime_df = anime_df.drop(bad_rows).reset_index(drop=True)
```

```python
anime_df.info()
```

```python
anime_df['score'] = pd.to_numeric(anime_df['score'], errors='coerce')
anime_df['scored_by'] = pd.to_numeric(anime_df['scored_by'], errors='coerce')
anime_df['members'] = pd.to_numeric(anime_df['members'], errors='coerce')
anime_df['favorites'] = pd.to_numeric(anime_df['favorites'], errors='coerce')
anime_df['rank'] = pd.to_numeric(anime_df['rank'], errors='coerce')
anime_df['anime_id'] = pd.to_numeric(anime_df['anime_id'], errors='coerce')
```

```python
columns=anime_df.columns
columns
```

```python
def printnas(df):
  for i in columns:
    print(df[i].isna().value_counts());
```

```python
anime_df = anime_df.dropna(subset=['synopsis'])
```

```python
anime_df = anime_df.dropna(subset=['type'])
```

```python
import json
json_cols = ['genres','explicit_genres','themes', 'demographics', 'studios']
```

```python
def json_counts(anime_df):
  for i in json_cols:
    print(anime_df[i].value_counts())
```

```python
json_counts(anime_df)
```

```python
anime_df['explicit_genres'].value_counts()
```

```python
for col in json_cols:
    anime_df[col] = anime_df[col].apply(lambda x: json.loads(x) if pd.notna(x) else [])
```

```python
anime_df = anime_df.drop(columns=['explicit_genres'])
```

```python
json_cols.remove('explicit_genres')
```

```python
json_counts(anime_df)
```

```python
anime_df['synopsis'].value_counts()
```

```python
placeholder = "No synopsis has been added for this series yet."
anime_df = anime_df[~anime_df['synopsis'].str.contains(placeholder, na=False)].reset_index(drop=True)

anime_df = anime_df[anime_df['synopsis'].str.len() > 100].reset_index(drop=True)
```

```python
anime_df.shape
```

```python
anime_df['type'].value_counts()
```

```python
anime_df['rating'].value_counts()
```

```python
anime_df = anime_df[~anime_df['type'].isin(['Music', 'CM', 'PV'])].reset_index(drop=True)
anime_df = anime_df[anime_df['rating'] != 'Rx - Hentai'].reset_index(drop=True)
anime_df.shape
```

```python
json_counts(anime_df)
```

```python
anime_df['tags'] = anime_df['genres'] + anime_df['themes'] + anime_df['demographics']
anime_df['tags'] = anime_df['tags'].apply(lambda x: ' '.join(x))
```

```python
anime_df['text_for_embedding'] = anime_df['synopsis'] + ' ' + anime_df['tags']
```
